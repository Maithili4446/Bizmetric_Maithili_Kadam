```python
cart = []

ask = input("Books(1) or Notebook(2): ")

# ---------------- BOOKS ----------------
if ask == '1':

    standard = int(input("Enter Standard: "))

    if 1 <= standard <= 4:
        std = '1-4'
    elif 5 <= standard <= 8:
        std = '5-8'
    else:
        std = '9-10'

    books = {
        '1-4': {'Hindi':60,'Marathi':60,'English':80,'Science':90,'Maths':100},
        '5-8': {'Hindi':100,'Marathi':100,'English':100,'Science':120,'Maths':140},
        '9-10': {'Hindi':150,'Marathi':150,'English':150,'Science':200,'Maths':250}
    }

    n = int(input("How many books: "))

    for i in range(n):
        subject = input("Enter Subject: ").title()
        qty = int(input("Quantity: "))

        price = books[std][subject]
        total = price * qty

        cart.append([subject, price, qty, total])


# ---------------- NOTEBOOK ----------------
else:

    pages = int(input("Pages (100/200): "))

    notebooks = {
        100: {'square':40,'4lines':30,'2lines':20,'single line':60,'A4':100},
        200: {'square':70,'4lines':50,'2lines':50,'single line':100,'A4':180}
    }

    n = int(input("How many notebooks: "))

    for i in range(n):
        book_type = input("Type: ").lower()
        qty = int(input("Quantity: "))

        price = notebooks[pages][book_type]
        total = price * qty

        cart.append([book_type, price, qty, total])


# ---------------- BILL ----------------
bill = input("Do you want bill? (yes/no): ")

if bill.lower() == 'yes':

    print("\n----------- BILL -----------")
    print("Sr\tName\t\tPrice\tQty\tTotal")

    grand_total = 0

    for i, item in enumerate(cart, 1):
        print(f"{i}\t{item[0]}\t\t{item[1]}\t{item[2]}\t{item[3]}")
        grand_total += item[3]

    print("----------------------------")
    print("Grand Total = ₹", grand_total)
    print("----------------------------")
```

```python
books_dict = {
    "Hindi": {
        "1st-4th": 60,
        "5th-8th": 100,
        "9-10th": 150
    },
    "Marathi": {
        "1st-4th": 60,
        "5th-8th": 100,
        "9-10th": 150
    },
    "English": {
        "1st-4th": 80,
        "5th-8th": 100,
        "9-10th": 150
    },
    "Science": {
        "1st-4th": 90,
        "5th-8th": 120,
        "9-10th": 200
    },
    "Maths": {
        "1st-4th": 100,
        "5th-8th": 140,
        "9-10th": 250
    }
}

notebook_dict = {
    "square": {100: 40, 200: 70},
    "4 lines": {100: 30, 200: 50},
    "2 lines": {100: 30, 200: 50},
    "single lines": {100: 60, 200: 100},
    "A4": {100: 100, 200: 180}
}

bill = []
total = 0

standard = int(input("Enter Standard: "))

if 1 <= standard <= 4:
    std_key = "1st-4th"
elif 5 <= standard <= 8:
    std_key = "5th-8th"
elif 9 <= standard <= 10:
    std_key = "9-10th"
else:
    print("Invalid Standard")
    exit()

choice = input("Do you want Books? (yes/no): ")

while choice == "yes":
    book = input("Enter Book Name: ")
    qty = int(input("Enter Quantity: "))
    price = books_dict[book][std_key]
    amount = price * qty
    bill.append((book, qty, amount))
    total += amount
    choice = input("Continue Books? (yes/no): ")

choice = input("Do you want Notebook? (yes/no): ")

while choice == "yes":
    n_type = input("Enter Notebook Type: ")
    pages = int(input("Enter Pages: "))
    qty = int(input("Enter Quantity: "))
    price = notebook_dict[n_type][pages]
    amount = price * qty
    bill.append((f"{n_type} Notebook {pages}", qty, amount))
    total += amount
    choice = input("Continue Notebook? (yes/no): ")

gen = input("Generate Bill? (yes/no): ")

if gen == "yes":
    print("\n------ BILL ------")
    for item in bill:
        print(item[0], "x", item[1], "=", item[2])
    print("Total Amount:", total)
else:
    print("Bill not generated")
```
