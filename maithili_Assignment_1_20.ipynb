12. Ask user to enter information about salary of the employee per year and rating received as A, B, C, D

If salary upto 5 lpa then increament based on ratings are A = 16% , B=12%, C= 10%, D=6%

If salary upto 10 lpa then increament based on ratings are A = 14% , B=10%, C= 8%, D=6%

If salary upto 15 lpa then increament based on ratings are A = 8% , B=6%, C= 4%, D = no increment

If salary upto 23 lpa then increament based on ratings are A = 7% , B=5%, C= 4%, D=no

```python
marks = int(input("Enter your marks: "))

if marks > 80:
    print("You got Distinction")
elif marks > 60:
    print("You got First Class")
elif marks > 40:
    print("You got Second Class")
elif marks >= 35:
    print("PASS")
else:
    print("FAIL")
```

```python
salary = float(input("Enter annual salary (in LPA): "))
rating = input("Enter rating (A/B/C/D): ")

increment = 0

if salary <= 5:
    if rating == 'A':
        increment = 16
    elif rating == 'B':
        increment = 12
    elif rating == 'C':
        increment = 10
    elif rating == 'D':
        increment = 6

elif salary <= 10:
    if rating == 'A':
        increment = 14
    elif rating == 'B':
        increment = 10
    elif rating == 'C':
        increment = 8
    elif rating == 'D':
        increment = 6

elif salary <= 15:
    if rating == 'A':
        increment = 8
    elif rating == 'B':
        increment = 6
    elif rating == 'C':
        increment = 4
    elif rating == 'D':
        increment = 0

elif salary <= 23:
    if rating == 'A':
        increment = 7
    elif rating == 'B':
        increment = 5
    elif rating == 'C':
        increment = 4
    elif rating == 'D':
        increment = 0

new_salary = salary + (salary * increment / 100)
print("Increment:", increment, "%")
print("New Salary:", new_salary, "LPA")
```

```python
password=input("Enter password: ")
# The 'rev' variable is not used for indexing, so it can be removed or left as is, but its content is irrelevant for the print statements.
# rev=" "
for i in range(0,len(password),2):
  # Calculate indices for the pair from the end of the password string.
  # These were the original indices, but incorrectly applied to 'rev'.
  idx1 = len(password)-(i+2)
  idx2 = len(password)-(i+1)

  # Ensure both indices are valid before printing a pair.
  if idx1 >= 0 and idx2 >= 0:
    print(password[idx1],end=" ")
    print(password[idx2],end=" ")
  # If the password has an odd length, the very first character (at index 0)
  # will be left after all pairs are processed. This condition handles that single character.
  elif idx2 == 0 and len(password) % 2 != 0:
    print(password[idx2],end=" ")
```
